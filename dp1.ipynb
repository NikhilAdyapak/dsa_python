0/1 Knapsack

```python
def knapsack(weight, value, W, n):
    if (W == 0) or (n == 0):
        return 0
    
    if (weight[n-1] <= W):
        return max(value[n-1] + knapsack(weight, value, W-weight[n-1], n-1), knapsack(weight, value, W, n-1))
    
    else:
        return knapsack(weight, value, W, n-1)
```

```python
W = 6
profit = [1, 2, 3]
weight = [4, 5, 1]
print(knapsack(weight, profit, W, len(weight)))
```

Memorize

```python
import numpy as np
arr = np.full((10,10), -1)
print(arr)
```

```python
def mem_knapsack(weight, value, W, n):
    if arr[n][W] != -1:
        return arr[n][W]
    
    if weight[n-1] <= W:
        arr[n][W] = max(value[n-1] + knapsack(weight, value, W-weight[n-1], n-1), 
                        knapsack(weight, value, W, n-1)) 
        return arr[n][W]
    else:
        arr[n][W] = knapsack(weight, value, W, n-1)
        return arr[n][W]
```

```python
print(mem_knapsack(weight, profit, W, len(weight)))
```

```python
# W = 7
weight = [1,3,4,7,2,5]
values = [21,45,6,7,34,5]
n = len(weight)
dp = np.zeros((n+1,W+1))
dp
```

```python
def dp_knapsack(weight, values, W, n):

    for i in range(n+1):
        for j in range(W+1):
            if (n == 0) or (W == 0):
                dp[i][j]

            pick = 0

            if weight[i-1] <= j:
                pick = values[i-1] + dp[i-1][j-weight[i-1]]

            not_pick = dp[i-1][j]

            dp[i][j] = max(pick, not_pick)
    return dp[n][W]
```

```python
def generate(W):
    dp = dp_knapsack(weight, values, W, n)
    print(dp)
```
